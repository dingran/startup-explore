```python
startup_data=pd.read_csv('overall_company_list_expanded.csv')
```

```python
feature_list=startup_data.columns.values[29:]
```

```python
features=startup_data[feature_list]
```

```python
features=sk.preprocessing.normalize(features)
```

```python
print feature_list[features[6]>0]
print startup_data.iloc[6]['Categories']
```

```python
kmeans_model=KMeans(n_clusters=8,n_init=20)
kmeans_model.fit(features)
centroids, cluster_assignment = kmeans_model.cluster_centers_, kmeans_model.labels_
```
